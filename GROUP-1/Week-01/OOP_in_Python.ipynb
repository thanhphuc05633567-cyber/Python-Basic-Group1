Class and Instance Attributes

```python
class car:
    # This acts like a constructor. It sets up the initial state.
    def __init__ (self, name ,brand , year): #Self is referencing the data itself

        self.name = name #setting up the name unique to the instance
        self.brand = brand
        self.year = year
    def sound(self): #Defining a method within the class
        return "Vroom Vroom!"
#Applying the instances:
carA = car('A', "Lamborghini" , 2024)
carB = car('B', "Honda", 2017)

#Accessing a certain data:

print(f"Car A is named {carA.name}, with the brand {carA.brand} and manufactured in {carA.year}")
print (f"Car B is named {carB.name}, with the brand {carB.brand} and manufactured in {carB.year}" )
#Both share the same class definition but holds different datas

#Implementation of the methods within the class:
print(carA.sound())
```

Inheritance and MRO

```python
print("INHERITANCE")

class Animal1:
    def eat(self):
        print("eating")

class Dog1(Animal1):
    def bark(self):
        print("Woof")

d1 = Dog1()
d1.eat()
d1.bark()

print("\nMETHOD OVERRIDING")

class Animal2:
    def sound(self):
        print("sound")

class Dog2(Animal2):
    def sound(self):
        print("Woof")

d2 = Dog2()
d2.sound()

print("\nUSING super()")

class Animal3:
    def sound(self):
        print("sound")

class Dog3(Animal3):
    def sound(self):
        super().sound()
        print("Woof")

d3 = Dog3()
d3.sound()

print("\nMULTIPLE INHERITANCE")

class Animal4:
    def sound(self):
        print("sound")

class Walker:
    def move(self):
        print("Walk")

class Swimmer:
    def move_in_water(self):
        print("Swim")

class Duck(Animal4, Walker, Swimmer):
    def sound(self):
        print("Quack")

duck = Duck()
duck.sound()
duck.move()
duck.move_in_water()

print("\nMETHOD RESOLUTION ORDER (MRO)")
print("\nSingle Inheritance:")
print(Dog3.__mro__)

print("\nMultiple Inheritance:")
print(Duck.__mro__)
```

Abstract Base Classes

```python
from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

class Circle(Shape):

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

    def perimeter(self):
        return 3.14 * self.radius * 2

class Square(Shape):
    pass # This does not implement abstract methods area() and perimeter()

# shape = Shape()  # This would raise TypeError
# square = Square() # This would raise TypeError

circle = Circle(5)
print(circle.area())
print(circle.perimeter())
```

Magic Methods

```python
# Magic Methods = Dunder methods (double underscore) __init__, __str__, __eq__ ,...
# automatically called by many of Python's built-in operations, allowing us to define or customize the behavior of objects

class Book:
    def __init__(self, author, title, pages_num):
        self.title = title
        self.author = author
        self.pages_num = pages_num

    def __str__(self):
        return f"{self.title} by {self.author}"

    def __eq__(self, other):
        return (self.title == other.title) and (self.author == other.author) #dont care about the num of pages

    def __lt__(self, other):
        return self.pages_num < other.pages_num

    def __gt__(self, other):
        return self.pages_num > other.pages_num

    def __add__(self, other):
        return self.pages_num + other.pages_num

    def __contains__(self, keyword):
        return keyword in self.title
    
    def __getitem__(self, key):
        if key == 'title':
            return self.title

        elif key == 'author':
            return self.author

        elif key == 'pages_num':
            return self.pages_num
        
        else:
            return f"{key} was not valid"

b1 = Book("Kelvin", "A", 150)
b2 = Book("John", "B", 200)
b3 = Book("Kathy", "C", 300)
b4 = Book("John", "B", 201)
# __str__()
print(b1)
# __eq__()
print(b1 == b2)
print(b2 == b4)
#__lt__()
print(b1 < b2)
print(b3 < b4)
# __gt__()
print(b1 > b2)
print(b3 > b4)
# __add__()
print(b1 + b2)
# __contains__()
print("John" in b4)
print("John" in b3)
# __getitem__()
print(b1['title'])
```
